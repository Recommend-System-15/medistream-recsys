```python
import pandas as pd
from konlpy.tag import *; okt = Okt()
from sklearn.feature_extraction.text import CountVectorizer
from tqdm.notebook import tqdm
from bertopic import BERTopic
```

```python
# 영수님이 만드신 파일 upload
df1 = pd.read_json('/home/user_1/medistream-recsys/Script/YS/df_book_clean.json')
df = df1.copy()
```

```python
contents_book = df['description']
```

```python
contents_book[0]
```

# Making Token

```python
p_contents = []
for doc in tqdm(contents_book):
    tokens = [token for token in okt.nouns(doc) if len(token) > 1]
    p_contents.append(" ".join(tokens))

# print(p_contents[1])
```

```python
p_contents
```

# Modeling

```python
vectorizer = CountVectorizer(max_features=3000)
model = BERTopic(embedding_model='sentence-transformers/distiluse-base-multilingual-cased-v1',
                 vectorizer_model=vectorizer,
                 nr_topics=10,
                 top_n_words=10,
                 calculate_probabilities=True)
```

```python
# topics, probs = model.fit(p_contents) # 책 수만큼
# model.fit(p_contents)
topics, probs = model.fit_transform(p_contents)
```

# Visualization
